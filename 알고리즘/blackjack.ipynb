## 문제
민재는 도박에 빠져 강원랜드에서 모든 재산을 탕진했다. 하지만 아직도 그 손맛을 잊지못해, 집에서 혼자 블랙잭 게임을 만들어 즐기려고 한다. 

다음과 같은 기능을 가진 블랙잭 게임을 제작하시오.

## 기능
블랙잭은 조커를 제외한 52장의 카드로 진행하는 게임이다. 

플레이어는 딜러와 플레이어가 진행하는 게임으로, 21에 가까운 숫자를 만드는 사람이 이기는 게임이다.
J,Q,K는 10으로 계산하며, A는 1로 사용할 수도 있고 11로 사용할 수 도 있다.

프로그램이 시작되면 플레이어는 1000원을 갖고 게임을 시작한다.

게임이 시작되면 먼저 배팅금액을 선택할 수 있는데, 100원 혹은 200원중 하나를 선택할 수 있다.
만약 배팅금액이 100원 이하라면, 게임은 종료된다.

배팅이 끝나면, 딜러는 한장 플레이어는 카드를 두장 받는다.

플레이어는 두가지 행동 중에 하나를 선택하는데, 힛 또는 스테이할 수 있다.
'힛' 하게 되면 카드를 한장 더 받게되고, '스테이'하면 턴이 종료된다. '힛'은 원하는 만큼 할 수 있지만, 만약 가지고있는 카드의 숫자가 21을 넘게 되면 '버스트'되어 배팅한 돈을 잃게 된다.

플레이어가 '스테이'를 하게 되면 딜러는 카드를 한장 더 받고 딜러의 턴이 진행된다. 딜러는 카드의 총합이 17점 미만이면 자동으로 무조건 '힛'해야 하며, 딜러의 카드 총합이 17점 이상일때는 플레이어와 숫자를 비교하여, 플레이어가 이겼다면 배팅한 돈의 두배를 받게되고, 플레이어가 지면 배팅한 돈 모두를 잃는다. 딜러도 마찬가지로 21이 넘으면 '버스트'되어 플레이어는 배팅한 돈의 두배를 가져가게 된다. 

만약 첫 패에 A한장과 J,Q,K,10중에 한장을 받아 한번에 21이 된다면, 이는 '블랙잭'이 되어 딜러의 패와 상관없이 딴돈의 1.5배를 받게된다.
